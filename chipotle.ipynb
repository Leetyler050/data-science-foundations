### Part 2: Separate `file_nested_list` into the `header` and the `data`.


```python
cdf['item_price_rm'] = cdf['item_price'].apply(lambda x: (str(x).replace("$", "")))
cdf.head()

#'{:f}'.format(3.141592653589793) example of float formatting 
```

```python
cdf['item_price_rm'] = cdf['item_price_rm'].apply(lambda x: float(x))
```
